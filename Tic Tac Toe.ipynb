```python
print('Gra w kółko Kółko i krzyżyk!', flush = True)
plansza = ['-','-','-','-','-','-','-','-','-']

def board():
    return print(f'Obecny stan planszy:\n{plansza[0]+plansza[1]+plansza[2]}\n{plansza[3]+plansza[4]+plansza[5]}\n{plansza[6]+plansza[7]+plansza[8]}', flush=True)

board()


print('Zaczynajmy zabawę! Drodzy gracze - przedstawcie się :)')
gracz_1 = input('Graczu nr 1 podaj swoje imię:')
print(f'Gracz 1: {gracz_1}', flush = True)
gracz_2 = input('Graczu nr 2 podaj swoje imię: ')
print(f'Gracz 2: {gracz_2}', flush = True)
print(f'{gracz_1} i {gracz_2} - niech wygra lepszy!', flush = True)

symbol_gr1 = input('Podaj symbol gracza nr 1 (wybierz X lub O: ').upper()
while symbol_gr1 != 'X' or symbol_gr1 != 'O':
    if symbol_gr1 == 'X':
        symbol_gr2 = 'O'
        print(f'Symbol gracza nr 1 to: {symbol_gr1}\nSymbol gracza nr 2 to: {symbol_gr2}')
        break
    elif symbol_gr1 == 'O':
        symbol_gr2 = 'X'
        print(f'Symbol gracza nr 1 to: {symbol_gr1}\nSymbol gracza nr 2 to: {symbol_gr2}')
        break
    else:
        print('Podałeś błędny symbol ! Wpisz X lub O')
        symbol_gr1 = input('Podaj symbol gracza nr 1 (wybierz X lub O: ').upper()

print('Grę zaczyna gracz nr 1!')

def wygrana():
    if (plansza[0]==plansza[1] and plansza[1] == plansza[2] and plansza[0] == symbol_gr1) or (plansza[0]==plansza[1] and plansza[1] == plansza[2] and plansza[0] == symbol_gr2) :
        return True
    elif (plansza[6]==plansza[7] and plansza[7] == plansza[8] and plansza[6] == symbol_gr1) or (plansza[6]==plansza[7] and plansza[7] == plansza[8] and plansza[6] == symbol_gr2) :
        return True
    elif (plansza[0]==plansza[3] and plansza[3] == plansza[6] and plansza[0] == symbol_gr1) or (plansza[0]==plansza[3] and plansza[3] == plansza[6] and plansza[0] == symbol_gr2) :
        return True
    elif (plansza[0]==plansza[4] and plansza[4] == plansza[8] and plansza[0] == symbol_gr1) or (plansza[0]==plansza[4] and plansza[4] == plansza[8] and plansza[0] == symbol_gr2):
        return True
    elif (plansza[1]==plansza[4] and plansza[4] == plansza[7] and plansza[1] == symbol_gr1) or (plansza[1]==plansza[4] and plansza[4] == plansza[7] and plansza[1] == symbol_gr2):
        return True
    elif (plansza[2]==plansza[4] and plansza[4] == plansza[6] and plansza[2] == symbol_gr1) or (plansza[2]==plansza[4] and plansza[4] == plansza[6] and plansza[2] == symbol_gr2):
        return True
    elif (plansza[2]==plansza[5] and plansza[5] == plansza[8] and plansza[2] == symbol_gr1) or (plansza[2]==plansza[5] and plansza[5] == plansza[8] and plansza[2] == symbol_gr2):
        return True
    elif (plansza[3]==plansza[4] and plansza[4] == plansza[5] and plansza[3] == symbol_gr1) or (plansza[3]==plansza[4] and plansza[4] == plansza[5] and plansza[3] == symbol_gr2):
        return True
    elif (plansza[6]==plansza[7] and plansza[7] == plansza[8] and plansza[6] == symbol_gr1) or (plansza[6]==plansza[7] and plansza[7] == plansza[8] and plansza[6] == symbol_gr2):
        return True
wygrana()

def rozgrywka():
    """
    Gra kółko i krzyżyk polega na ułożeniu symboli X lub O w sekwencji 3 w rzędzie:
    - w pionie lub
    - w poziomie lub
    - na ukos.
    Wygrywa ten gracz, który pierwszy ułoży swoje symbole w podanej sekwencji.
    
    W przypadku braku wolnych pól mamy remis - zalecamy powtórzyć rozgrywkę :)
    
    Wyjaśnienie do gry:
    Każde pole ma swój numer, który należy wprowadzić aby postawić symbol w odpowiednim miejscu.
    Licząc od lewego górnego rogu pole ma numer 1, kolejne 2 itd jak rozpisane poniżej:
    1 2 3
    4 5 6
    7 8 9

    Przykładowe ułożenia symboli 'wygrywające':

    - X -
    - X -
    - X -

    lub

    O - -
    - O -
    - - O
    
    Udanej zabawy!"""
    board()
    while wygrana != True:
            def gdzie_gr1():
                ruch_gr1 = int(input('Graczu nr 1, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
                zakres = False
                while ruch_gr1 not in range(1,10):
                    print('Podałeś liczbę z poza zakresu!')
                    ruch_gr1 = int(input('Graczu nr 1, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
                puste = True
                global plansza
                while puste:
                    if plansza[ruch_gr1-1] != 'X' and plansza[ruch_gr1-1] != 'O':
                        plansza[ruch_gr1-1] = symbol_gr1
                        return
                    else:
                        print('To pole jest zajęte!')
                        ruch_gr1 = int(input('Graczu nr 1, gdzie chcesz postawić swój symbol? Poprzednio wybrane pole jest zajęte ! Podaj liczbę od 1 do 9:'))    
                        while ruch_gr1 not in range(1,10):
                            print('Podałeś liczbę z poza zakresu!')
                            ruch_gr1 = int(input('Graczu nr 1, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
            gdzie_gr1()
            board()
            if plansza.count('X') + plansza.count('O') == 9:
                print('Brak wolnych pól. Remis. Zagrajcie jeszcze raz :)')
                break
            else:
                wygrana()
                if wygrana() == True:
                    print(f'Koniec gry! Wygrał {gracz_1} ! Gratuluję, jesteś thebeściak')
                    break
            def gdzie_gr2():
                ruch_gr2 = int(input('Graczu nr 2, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
                zakres = False
                while ruch_gr2 not in range(1,10):
                    print('Podałeś liczbę z poza zakresu!')
                    ruch_gr2 = int(input('Graczu nr 2, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
                puste = True
                global plansza
                while puste:
                    if plansza[ruch_gr2-1] != 'X' and plansza[ruch_gr2-1] != 'O':
                        plansza[ruch_gr2-1] = symbol_gr2
                        return
                    else:
                        print('To pole jest zajęte!')
                        ruch_gr2 = int(input('Graczu nr 2, gdzie chcesz postawić swój symbol? Poprzednio wybrane pole jest zajęte ! Podaj liczbę od 1 do 9:'))    
                        while ruch_gr2 not in range(1,10):
                            print('Podałeś liczbę z poza zakresu!')
                            ruch_gr2 = int(input('Graczu nr 2, gdzie chcesz postawić swój symbol? Podaj liczbę od 1 do 9:'))
            gdzie_gr2()
            board()
            if plansza.count('X') + plansza.count('O') == 9:
                print('Brak wolnych pól. Remis. Zagrajcie jeszcze raz :)')
                break
            else:
                wygrana()
                if wygrana() == True:
                    print(f'Koniec gry! Wygrał {gracz_2} ! Gratuluję, jesteś thebeściak')
                    break

rozgrywka()
```
