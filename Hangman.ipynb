```python
def hangman():
  print('Game: Hangman', flush = True)
  print("""
  _                                             
  | |                                            
  | |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
  | '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
  | | | | (_| | | | | (_| | | | | | | (_| | | | |
  |_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                      __/ |                      
                    |___/ )
  """)

  import random

  stages = ['''
    +---+
    |   |
    O   |
  /|\  |
  / \  |
        |
  =========
  ''', '''
    +---+
    |   |
    O   |
  /|\  |
  /    |
        |
  =========
  ''', '''
    +---+
    |   |
    O   |
  /|\  |
        |
        |
  =========
  ''', '''
    +---+
    |   |
    O   |
  /|   |
        |
        |
  =========''', '''
    +---+
    |   |
    O   |
    |   |
        |
        |
  =========
  ''', '''
    +---+
    |   |
    O   |
        |
        |
        |
  =========
  ''', '''
    +---+
    |   |
        |
        |
        |
        |
  =========
  ''']

  end_of_game = False

  chosen_word = input('What is hidden word?')
  word_length = len(chosen_word)

  lives = 6

  #Testing code
  #print(f'Pssst, the solution is {chosen_word}.') # That part is just to check if game is running properly. SHall be turned off for real game.

  #Create blanks
  display = []
  for _ in range(word_length):
      display += "_"

  guessed_letters =[]
  while not end_of_game:
      guess = input("Guess a letter: ").lower()
      guessed_letters.append(guess)
      if guess in display:
          print(f"You've already guessed {guess}", flush = True)
          print(f'List of chosen letters: {guessed_letters}', flush = True)

      #Check guessed letter
      for position in range(word_length):
          letter = chosen_word[position]
          if letter == guess:
              display[position] = letter
              

      if guess not in chosen_word:
          print(f"You guessed {guess}, that's not the hidden word. You lose a life.", flush = True)
          print(f'List of chosen letters: {guessed_letters}', flush = True)
          lives -= 1
          if lives == 0:
              end_of_game = True
              print("You lose.", flush = True)
              print(f"Hidden worrd is {chosen_word}", flush=True)

      #Join all the elements in the list and turn it into a String.
      print(f"{' '.join(display)}", flush=True)

      #Check if user has got all letters.
      if "_" not in display:
          end_of_game = True
          print("You win.", flush = True)

      print(stages[lives])
```

```python
hangman()
```
