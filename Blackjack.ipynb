```python
rules = """############### Our Blackjack House Rules #####################

## The deck is unlimited in size. 
## There are no jokers. 
## The Jack/Queen/King all count as 10.
## The the Ace can count as 11 or 1.
## Use the following list as the deck of cards:
## cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
## The cards in the list have equal probability of being drawn.
## Cards are not removed from the deck as they are drawn.
## The computer is the dealer."""

logo = """
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
"""

import random

def game_blackjack():

    print(logo)
    print(rules)
    print('-----------------------------------------------------------------------', flush = True)

    def deal_card():

        cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

        return random.choice(cards)

    i=0
    user_cards = []
    computer_cards = []

    #Here we define random cards for user and computer
    while i < 2 :
        user_cards.append(deal_card())
        computer_cards.append(deal_card())
        i = i +1

    user_sum = sum(user_cards)
    computer_sum = sum(computer_cards)

    # 0 will represent a blackjack in our game
    blackjack = 21

    if user_sum == blackjack:
        user_sum = 0

    if computer_sum == blackjack:
        computer_sum = 0

    #If the score is already over 21, remove the 11 and replace it with a 1.
    if 11 in user_cards and user_sum > 21:
        user_cards.remove(11)
        user_cards.append(1)
        user_sum = sum(user_cards)

    if 11 in computer_cards and computer_sum > 21:
        computer_cards.remove(11)
        computer_cards.append(1)
        computer_sum = sum(computer_cards)

    
        
    def calculate_score():

        user_sum = sum(user_cards)
        computer_sum = sum(computer_cards)

        if user_sum == blackjack:
            user_sum = 0

        if computer_sum == blackjack:
            computer_sum = 0

        #If the score is already over 21, remove the 11 and replace it with a 1.
        if 11 in user_cards and user_sum > 21:
            user_cards.remove(11)
            user_cards.append(1)
            user_sum = sum(user_cards)

        if 11 in computer_cards and computer_sum > 21:
            computer_cards.remove(11)
            computer_cards.append(1)
            computer_sum = sum(computer_cards)

        print(f'Current user cards are {user_cards}', flush = True)
        print(f'Computer first card is {computer_cards[0]}', flush = True)    
        print('-----------------------------------------------------------------------', flush = True)
        print(f"Current user sum is: {user_sum}", flush = True)
        print('-----------------------------------------------------------------------', flush = True)

        if user_sum == 0 or  computer_sum == 0 or user_sum > 21:
            print('Game is finished', flush=True)
            print(f'Current user cards are {user_cards} and computer cards are {computer_cards}', flush = True)
            print('-----------------------------------------------------------------------', flush = True)
            print(f"User sum is: {user_sum} and computer sum is {computer_sum}", flush = True)
            if user_sum == computer_sum:
                print('It is a draw') 
            elif computer_sum == 0:
                print('You lost')
            elif user_sum == 0:
                print('You won')
            elif user_sum > 21:
                print('You lost')
            elif computer_sum > 21:
                print('You won')
            elif user_sum > computer_sum:
                print('You won')
            else:
                print('You lost')
                
        else:
            next_card = input('Do you would like to draw the next card? Select y or n')
            if next_card == 'y':
                user_cards.append(deal_card())
                calculate_score()
                                
            elif next_card =='n' and user_sum < 21:
                while computer_sum !=0 and computer_sum < 17:
                    computer_cards.append(deal_card())
                    computer_sum = sum(computer_cards)
                    if computer_sum == blackjack:
                        computer_sum = 0

                    if 11 in computer_cards and computer_sum > 21:
                        computer_cards.remove(11)
                        computer_cards.append(1)
                        computer_sum = sum(computer_cards)

                        print(f'Current user cards are {user_cards} and computer cards are {computer_cards}', flush = True)    
                        print('-----------------------------------------------------------------------', flush = True)
                        print(f"Current user sum is: {user_sum} and computer sum is {computer_sum}", flush = True)
                        print('-----------------------------------------------------------------------', flush = True)
                else:
                    print(f'Current user cards are {user_cards} and computer cards are {computer_cards}', flush = True)
                    print('-----------------------------------------------------------------------', flush = True)
                    print(f"Current user sum is: {user_sum} and computer sum is {computer_sum}", flush = True)
                    print('-----------------------------------------------------------------------', flush = True)
                    print('Game is finished', flush=True)
                    if user_sum == computer_sum:
                        print('It is a draw') 
                    elif computer_sum == 0:
                        print('You lost')
                    elif user_sum == 0:
                        print('You won')
                    elif user_sum > 21:
                        print('You lost')
                    elif computer_sum > 21:
                        print('You won')
                    elif user_sum > computer_sum:
                        print('You won')
                    else:
                        print('You lost')
                           
    return calculate_score()


def play():
    if input("Do you would like to play Blackjack? If yes put 'y', if no, then 'n'.") == 'y':
        game_blackjack()
    else:
        print('We can finish now.')

play()
```
