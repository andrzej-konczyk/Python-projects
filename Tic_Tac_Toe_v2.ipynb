```python

def print_board(board):
    print("-" * 13)
    for i in range(3):
        print("|", board[i * 3 + 0], "|", board[i * 3 + 1], "|", board[i * 3 + 2], "|")
        print("-" * 13)


def read_position(player):
    print(player, " na ruchu")
    position = int(input("Podaj pozycję od 1 do 9: "))
    return position - 1


def check_position(board, position):
    if board[position] == " ":
        return True
    else:
        return False


def check_win(board):
    win_positions = [
        [0, 1, 2],
        [3, 4, 5],
        [6, 7, 8],
        [0, 3, 6],
        [1, 4, 7],
        [2, 5, 8],
        [0, 4, 8],
        [2, 4, 6]
    ]

    for i in win_positions:
        if board[i[0]] == board[i[1]] == board[i[2]] != " ":
            return board[i[0]]

    return False


def main():
    board = [" ", " ", " ",
             " ", " ", " ",
             " ", " ", " "]

    win = False
    player = "O"
    while not win:
        print_board(board)
        position = read_position(player)
        while not check_position(board, position):
            position = read_position(player)
        board[position] = player
        win = check_win(board)
        if win:
            print_board(board)
            print("Wygrał gracz", player)
        else:
            if " " not in board:
                print_board(board)
                print("Remis")
                break
            if player == "O":
                player = "X"
            else:
                player = "O"


main()
```
